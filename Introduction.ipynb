### Basic mathematical operation in python using hydrogen-like functions

#### Theory
In this introductory exercise, we will "play" with the hydrogen-like wave function, solution of the non-relativistic Schrdinger equation for the hydrogen atom.
The hydrogen wave function is given by a product of a radial and an angular part:

$$\Psi_{nlm} (|n,l,m> \text{in Dirac's notation)} = R_{nl}(r)Y_{lm}(\theta, \phi)$$

$Y_{lm}(\theta,\phi)$ represents the spherical harmonic functions while $R_{nl}(r)$ represents the radial function which, in atomic units, reads:

$$R_{nl}(r)=(\frac{2Z}{n}^3\frac{(n-l-1)!}{2n(n+l)!})^{\frac{1}{2}}e^{\frac{Zr}{n}}(\frac{2Zr}{n})^lL_{n-l-1}^{2l+1}(\frac{2Zr}{n})$$

where $L_{a}^{b}(x)$ is the associated Laguerre function. Recall that $\Psi$ is a complex function and that the probability density is given by the square of the norm $|\Psi|^2$.

If we want to solve the Schrdinger equation for multi-electron system (cf the 3 body problem) approximations have to be made.
For multi-electronic atoms, a rough but qualitative approximation is the shielding effect or effective nuclear charge approximation. Here the net force experienced by the electron of interest is approximated as a "shielded" nuclear charge. This is taken into account by writing the nuclear charge as an effective nuclear charge $Z_{eff}$, where $Z_{eff}=Z-\sigma$ with $\sigma$ representing the shielding factor due to the coulomb repulsion of the remaining $N-1$ electrons.
 For example, the wave function associated to the $2s$ electron of the C atom can be approximated by $\Psi_{nlm}$ with $Z=Z_{eff}=3.22$ This approximation allows us to pick specific shells of a given atom from the periodic table (see Figure 1.1.2.3 in the following link: https://chem.libretexts.org/Courses/Saint_Marys_College_Notre_Dame_IN/CHEM_342%3A_Bio-inorganic_Chemistry/Readings/Week_1%3A_Analysis_of_Periodic_Trends/1.1%3A_Concepts_and_principles_that_explain_periodic_trends/1.1.2%3A_Effective_Nuclear_Charge)

 The eigenfunction of the hydrogen-like Hamiltonian are complex 3 dimensional functions that characterize the electron commonly called orbitals. Chemists usually use the $s,p,d,f,g,...$ notation where the letters correspond to the $l$ quantum number, it has the following correspondence to Dirac's notation:

$$|n,0,0>    = ns$$
$$|n,1,\pm1> = np_{\pm1}$$
$$|n,1,0>    = np_{0}$$
$$|n,2,0>    = nd_{0}$$
$$|n,2,\pm1> = nd_{\pm1}$$
$$|n,2,\pm2> = nd_{\pm2}$$

Expressing the orbitals as real functions leads to the common chemists notation $np_{x}, np_{y}, nd_{xy},$ etc 

#### 1d) Write a python functions that returns $\Psi_{nlm}$, $R_{nl}$ and $Y_{ml}$ for given value of $n,l,m, \theta,\phi$, $r$ and $Z_{eff}$. <br>
  
Use the spherical harmonics function "scipy.special.sph(m,l,theta,phi)" and associated Laguerre polynomials "scipy.special.assoc_laguerre(x,a,b)". Be careful with the definition of $\phi$ and $\theta$. In python they are defined as the polar and azimuthal angle, respectively. The opposite convention is also often seen.

```python
 def Psi(n,l,m,r,phi,thet,Z):
    '''
    n: principal quantum number
    l: azimuthal quantum number
    m: magnetic quantum number
    r, phi and thet: spherical coordinates
    Z: nuclear charge
    
    return the total wave function Psi, the radial part R and the angular part Y
    '''

    # angular part -> spherical harmonics
    Y = special.sph_harm(m,l,thet,phi)

    # radial part containing the associated Laguerre polynomials
    R = np.sqrt(((2*Z/n)**3)*((math.factorial(n-l-1))/(2*n*math.factorial(n+l))))
    R *= ((2*r*Z)/n)**l       #The *= multiplies the previous R by the factor on the right and reassigns the product to R 
    R *= np.exp(-Z*r/n)
    x = (2*r*Z)/n
    R *= special.assoc_laguerre(x,n-l-1,2*l+1)

    psi = Y*R

    return psi, R, Y
```

In order to perform the integral, you can either write your own integration function using, for example, the trapezoid rules:
$$\int_{x_0}^{x_f} \int_{y_0}^{y_f} \int_{z0}^{z_f} f(x,y,z) dxdydz =
$$ $$\Delta x \Delta y \Delta z \sum_i^{ni-1}\sum_j^{nj-1}\sum_k^{nk-1}f\big(x_0+\frac{1}{2}\Delta x +i*\Delta x,\, y_0+\frac{1}{2}*\Delta y+j*\Delta y,\, z_0+\frac{1}{2}\Delta z +k*\Delta z \big)$$
where $ni, nj$ and $nk$ are the number of integration points, and the integration step $\Delta a$ is given by $\Delta a=(a_f-a_0)/n_a$. Both defining the integration grid. <br>
The function could look like this:

```python
# numerical triple integration using the trapezoide rule
def triple_int(nr,nt,nph,r2,intD,*arg):
    '''
    nr, nt, nph  : nbr of integration points for r, theta and phi, respectively
    intD : function for the integrant with intD(phi,theta,r,*arg)
    r2 : 0 -> rf is the boundary for r
    *arg : list of additional arguments to pass to intD
    '''
    
    # integration limits
    r1 = 0.
    t1 = 0.
    t2 = 2*(np.pi) 
    p1 = 0.
    p2 = np.pi
    
    # integration steps
    dr = (r2-r1)/nr
    dt = (t2-t1)/nt
    dp = (p2-p1)/nph  
    
    ans = 0.
    for i in range(nr-1):
        for j in range(nt-1):
            for k in range(nph-1):
                ri = r1 + dr/2 + i*dr
                tj = t1 + dt/2 + j*dt
                pk = p1 + dp/2 + k*dp
                ans += intD(pk,tj,ri,*arg)
    ans *= (dr*dt*dp)
    
    return ans
```

Alternatively, you can use python's built-in function for triple integration. The syntax for the latter is:

$$integrate.tplquad(func, r1, r2, t1, t2, p1, p2)$$

where $r1,t1,p1$ and $r2,t2,p2$ are the boundaries of the integral. The integration steps are automaticaly calculated by the function. The function returns the result and an estimation of the numerical error. Additional information can be found here: https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.tplquad.html. \\

Here is an example on how to use python's build-in triple integration algorithm. You can play with the integration steps of our $triple\_int$ code and see how it infuences the result and how it compares to python's function.

```python
# perform the triple integral of r^2 over spherical coordinates r, theta, phi
# the analytical results is 2pi/3

# define integrant
def func(phi,theta,r):
    ans=r**2*np.sin(phi)
    return ans

#perform integration over the specified boundaries 
sol1 = integrate.tplquad(func, 0, 1, 0, 2*np.pi,0,np.pi)
sol2 = triple_int(150,150,100,1,func)
print(sol1)
print(sol2)
```

#### Transition dipole moment: Theory
A transition moment is a matrix element between an initial and final state: $<\Psi_f|\hat{o}|\Psi_i> = \int \int \int \Psi_f^* \ \hat{o} \  \Psi_i \ r^2 sin(\phi) dr d\phi d\theta$. For a dipole transition moment $\hat{o}=\hat{\textbf{r}}$. The $\hat{\textbf{r}}$ operator in cartesian coordinate is simply a vector $\textbf{r}=(r_x, r_y, r_z)$ and can be expressed in spherical coordinates:
$$r_z = rcos(\phi) $$
$$r_y = rsin(\phi)sin(\theta)$$
$$r_x = rsin(\phi)cos(\theta)$$

Let us go back to the hydrogen atom. We want to calculate the transition dipole moment between the $1s$ and $2p_0$ state of hydrogen. Let us only calculate the transition dipole moment for the $z$ component of the field, hence $\hat{o}=\hat{r}_{z}=rcos(\phi)$. The integral we need to calculate is 
$$\int_{0}^{2\pi} \int_{0}^{\pi} \int_{0}^{rf} \Psi_f^* \  r cos(\phi) \ \Psi_i \, r^2 sin(\phi) dr d\phi d\theta \, $$ 
where $\Psi_{f}$ is our final state (the $2p_0$ orbital of hydrogen) and $\Psi_{i}$ is our initial state which corresponds to the $1s$ orbital of hydrogen. Our final integral becomes:
$$ \int_{0}^{2\pi} \int_{0}^{\pi} \int_{0}^{rf} \Psi_{2p_0}^* \, rcos(\phi) \, \Psi_{1s} \, r^2 sin(\phi) \, dr d\phi d\theta \,  $$ 

Python's integration function is not capable of dealing with complex numbers. Therefore we have to calculate the integrant for the real and imaginary part separately and sum the results.

The real or imaginary part of a complex number can be called using the following syntax:


```python
z = complex(4,2) # 4+2j
print(z)
print(z.real)
print(z.imag)

z=special.sph_harm(-1,2,np.pi/5,np.pi/3)
print(z)
print(z.real)
print(z.imag)
```

#### 1f) Write down a function that returns the real part of the integrant for the transition dipole matrix element and an other one that returns its imaginary part.

```python
def int_tdm(phi, theta, r, real=True):
    """
    Calculate the transition dipole matrix element integral
    and return either the real or imaginary part based on `real`.
    """
    ans  = np.conj(H_2p(r,phi,theta, 0))
    ans *= (r * np.cos(phi))
    ans *= H_1s(r,phi,theta)
    ans *= (r**2 * np.sin(phi))

    if real:
      return ans.real
    else:
      return ans.imag


def int_tdm_real(phi,theta,r):
    return int_tdm(phi, theta, r, real=True)

def int_tdm_imag(phi,theta,r):
    return int_tdm(phi, theta, r, real=False)
```
