```python
def plot_plane_wave(k, theta):
    '''
    k: wave number
    theta: angle of the wave vector
    '''
    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)
    r = np.cos(theta)*X + np.sin(theta)*Y
    Z = np.cos(k*(r))
    plt.contourf(X, Y, Z)
    
    plt.show()

plot_plane_wave(1, np.pi/4)
```
